# Analysis of results

## Particle size (Characteristic function)

```python
cha_index = ["frac_b1", "psize_Bronze_1", "psize_Bronze_2"]
CHA = val.loc[cha_index, :]
CHA
```

```python
import diffpy.srfit.pdf.characteristicfunctions as F


def cha_func(x, params):
    a = params.frac_b1
    f1 = F.sphericalCF(x, params.psize_Bronze_1)
    f2 = F.sphericalCF(x, params.psize_Bronze_2)
    return a * f1 + (1 - a) * f2


def draw_function(df_cha, df_data):
    plt.xlabel(r"r ($\AA$)")
    plt.ylabel(r"f ($\AA^{-2}$)")

    datalist = []
    for sample, params in df_cha.iteritems():
        data = {}
        data["sample"] = sample
        data["x"] = x = np.arange(0., 60.001, 0.1)
        data["y"] = y = cha_func(x, params)
        data["x0"] = x[np.argmax(y < 1e-12)]
        data["color"] = df_data.loc[sample, 'color']
        datalist.append(data)

    datalist = sorted(datalist, key = lambda data: data["x0"])

    for data in datalist:
        plt.plot(data["x"], data["y"], label=data["sample"], c=data["color"])
        plt.axvline(data["x0"], ls="--", c=data["color"])

    plt.legend(loc=3)
    H.config_ax()
    return
```

```python
plt.figure()
draw_function(CHA, DATA)
plt.show()
```

# Lattice Constants

```python
LAT = val.loc[["a_Bronze", "b_Bronze", "c_Bronze"]]
LAT
```

```python
def compare_lattice(df_lat):
    axs = []
    for n, (param, data) in enumerate(df_lat.iterrows()):
        ax = plt.axes([0.2, 0.6 - n * 0.2, 0.6, 0.2])
        ax.plot(data, "o", ms=10.)

        mean = data.mean()
        ax.axhline(mean, ls="--")
        ax.set_ylim(mean - 0.01, mean + 0.01)

        lat = param.split("_")[0]
        ylabel = rf"{lat} ($\AA$)"
        ax.set_ylabel(ylabel)
        H.config_ax()
        axs.append(ax)

    for ax in axs[:-1]:
        ax.set_xticklabels([])
    return
```

```python
plt.figure()
compare_lattice(LAT)
plt.show()
```
